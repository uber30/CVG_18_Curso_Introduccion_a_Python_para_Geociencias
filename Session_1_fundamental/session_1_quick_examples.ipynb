```python
# One-curve plot

# Importing an entire PEL
import numpy as np
# Importing a module PEL
import matplotlib.pyplot as plt

x = np.arange(1,101,1)

y = x**2 + 2*x + 2

plt.plot(x,y)
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
# Two-curve plot with grid, labels, and title

# libraries
import matplotlib.pyplot as plt
import numpy as np

# data for plotting
t = np.arange(0.0, 2.01, 0.01)
s = 1 + np.sin(2 * np.pi * t)
r = 1 + np.cos(2 * np.pi * t)

# Procedurla style plot
plt.plot(t, s)
plt.plot(t, r)
plt.xlabel('x')
plt.ylabel('s and r')
plt.title('Two waves')
plt.grid()
plt.show()
```

The previous plot is constructed using a **Procedural** programming style. The following plot is identical to the previous but done wit **Object-Oriented** programming style.

```python
# Two-curve plot with grid, labels, and title

# libraries
import matplotlib.pyplot as plt
import numpy as np

# data for plotting
t = np.arange(0.0, 2.01, 0.01)
s = 1 + np.sin(2 * np.pi * t)
r = 1 + np.cos(2 * np.pi * t)

# Object oriented style plot
fig, ax = plt.subplots()
ax.plot(t, s)
ax.plot(t, r)
ax.set_xlabel('x')
ax.set_ylabel('s and r')
ax.set_title('Two waves')
ax.grid(True)
plt.show()
```
