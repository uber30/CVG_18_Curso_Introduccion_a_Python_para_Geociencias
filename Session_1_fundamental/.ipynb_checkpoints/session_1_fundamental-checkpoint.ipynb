## Python External Libraries (PEL) functions

```python
# One-curve plot

# Importing an entire PEL
import numpy as np
# Importing a module PEL
import matplotlib.pyplot as plt

x = np.arange(1,101,1)

y = x**2 + 2*x + 2

plt.plot(x,y)
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
# Two-curve plot with grid, labels, and title

# libraries
import matplotlib.pyplot as plt
import numpy as np

# data for plotting
t = np.arange(0.0, 2.01, 0.01)
s = 1 + np.sin(2 * np.pi * t)
r = 1 + np.cos(2 * np.pi * t)

# plot
plt.plot(t, s)
plt.plot(t, r)
plt.xlabel('x')
plt.ylabel('s and r')
plt.title('Two waves')
plt.grid()
plt.show()
```

## Variables

In Python, variables point to reserved memory locations that store values. Along the execution of a program, the variables can have any type (strings, numbers, boolean ...), which are verified only during the execution of an operation or function. This behavior, known as **dynamic typing**, gives Python a lot of flexibility when declaring variables. However, in the end, the variables must agree with the operations you want to perform. In the examples below the operator `+` and `*` are examples of **overloading operators** which can perform different operations depending on the type of inputs:

```python
# Variables a and b are integers

a = 2
b = 3
```

```python
a + b
```

```python
a*5
```

```python
# Now they are going to be sequences of strings

a = 'These are'
b = ' dynamic typing and overloading operators'
```

```python
# + is an overloading operator

a + b
```

```python
# * is an overloading operator
b*3
```

### Variable names

* Variable names must **start with a letter or the underscore character**, not a number.  
* A variable name can only **contain alpha-numeric characters, and underscores** (A-z, 0-9, and _ ), not special characters. 
* Variable names are **case-sensitive** therefore python, Python, and PYTHON would be different variables.  

### Assigning values to variables

The equal sign opperator `=` assigns a value to a variable.

```python
# Simple calculation of the area of a circle inside the print()

pi = 3.1416
radius = 100
unit = 'square meters'

print('Area of the circle:', pi*(radius**2), unit)
```

```python
# A variable with the hypotenuse of a right triangle inside the print()

a = 30
b = 40
c = (a**2 + b**2)**0.5

unit = 'cm'  # use quotations to define a text

print("Hypotenuse:", c, unit)
```

## Basic data types

### Numeric data types

In Python, numeric data types represent data which have a numerical value
    
* **Integer**: A positive or negative whole number. Its limit depends on the memory size of our computer (32 or 64 bits).
* **Float**: A real number with floating point representation. It is specified by a decimal point.
* **Complex**: A number that is composed of a real part plus an imaginary part, j in the case of Python.

#### Assign variables

```python
i = 5
f = 6.0
comp = 7j
comp2 = 8.0 + 9.5j

print("Integer:", i)
print("Float or real:", f)
print("Complex:", comp)
print("Real + complex:", comp2)
```

```python
print(type(i))
print(type(f))
print(type(comp))
print(type(comp2))
```

#### Arithmetic operations can involve different numeric types

```python
# Sum of int and float

s = i + f
s
```

```python
# Sum of int and complex

s = i + comp
s
```

## Sequential data types

In Python a sequence is a collection of variables, generally of the same type. The most important sequences are string, sequence of strings (text), **list**, set, and **tuples**. Using an index we can access each element of the sequence, but be aware that **Python uses a zero-based numbering system**, which means that the initial item or element of a sequence has an index of 0, rather than 1. Therefore indexes range from 0 to n-1, where n is the number of items in the sequence:

    Sequence:  Python is fun and clever!
 
    Index:     0123456789 ...
 
 
 The function `len()` gives the length of the sequence

### List

A list is a **mutable sequence of different types**, inside square brackets and separated by commas.

```python
# Example

list1 = ["apple", "orange", 5, 7, True, None]
print(list1)
```

We can access one element of the list using the index between square brackets, but remember that the initial element of the list has an index 0.

```python
# Firts element

list1[0]
```

```python
# Other elements

print(list1[1])
print(list1[3])
list1[4]
```

We can also create a list as follows:

* An empty list: `list1=[]`
* A list with n equal values: `list1=[value]*n`

```python
# Empty list

list1 = []
print(list1)
```

```python
# List of 10 elements

list2 = [5]*10
print(list2)

len(list2)
```

As **lists are mutable**, we can change any element of the list (this is not the case for strings or tuples).

```python
# Changing 4th & 6th elements

list2[3] = 400
list2[5] = 600

print(list2)
```

#### Lists operations

Here are some useful operations for lists, performed by functions or methods:

### Tuple

A tuple is a **immutable sequence of different objects** separated by commas and inside parentheses (optional).

```python
# Example

tuple1 = "USA", "China", "Brazil"

print(tuple1, 'is a', type(tuple1))
print(tuple1[0], '\n')


tuple2 = ("AAA", 123, 456, 789, 'BBB')

print(tuple2, 'is also a', type(tuple2))
print(tuple2[4])
```

### List vs. tuples

Lists and tuples have subtle but important differences. Once created, lists can be modified, tuples on the other hand are immutable.

```python
# Example

list1 = [1,2,3,4]
tuple1 = (1,2,3,4)

print(list1)
print(tuple1)
```

```python
# Let's change the 3rd element in the list

list1[3] = 10
print(list1)
```

## Generating NumPy arrays

The are different ways to create NumPy arrays, some of the most important are:
    
* Converting a list or list of lists to an array with the function `np.array()`
* Creating a 1D array (vector) or 2D array (matrix) of zeros with the function `np.zeros()`
* Creating a 1D array or 2D array of fixed values with the function `np.full()`
* Creating a 1D array or 2D array of random values (floats) with the function `np.random.rand()`
* Creating a 1D array or 2D array of random values (integer) with the function `np.random.randint()`

Let's review them:

### List or list of lists to array

```python
%who
```

```python
%who module
```

#### List to array (vector)

```python
list1 = [0, 1, 2, 3]

vector = np.array(list1) 

print("List:", list1)
print(type(list1), '\n')

print("Vector:", vector)
print(type(vector))
vector.shape
```

#### List of lists to array (matrix)

```python
list_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

matrix = np.array(list_list)

print("List of lists:", list_list,'\n')
print("Matrix:\n", '\n', matrix)
matrix.shape
```

### Array of zeros

```python
vector = np.zeros(5)

print("Vector of zeros:", vector)
print(vector.dtype)
vector.shape
```

```python
vector + 5
```

```python
mat_zeros = np.zeros((5, 3))

print("Matrix of zeros:\n", '\n', mat_zeros)
mat_zeros.shape
```

### Array with constant values (full)

```python
vector5 = np.full(5, 10)

print("Vector with constant value:\n", '\n', vector5)
vector5.shape
```

```python
matrix5x5 = np.full((5, 5), 10)

print("Matrix with constant value:\n", '\n',matrix5x5)
matrix5x5.shape
```

### Array with random floats (0-1)

```python
vector_random = np.random.rand(5)

print("Vector with random values:", vector_random)
vector_random.shape
```

```python
matrix_random = np.random.rand(5, 3)

print("Matrix with random values:\n", '\n', matrix_random)
matrix_random.shape
```

```python
# Let's see the shape of the distribution of random numbers with a uniform distribution

import matplotlib.pyplot as plt

matrix_random = np.random.rand(100, 100)

plt.hist(matrix_random.flatten(), bins=50);
```

```python
# Let's see the shape of the distribution of random numbers with an standard normal distribution

matrix_random_2 = np.random.randn(100, 100)

plt.hist(matrix_random_2.flatten(), bins=50);
```

### Array with random integers

```python
random_integers = np.random.randint(50, size=(10))

print("Vector with random integers:\n", '\n', random_integers)
random_integers.shape
```
