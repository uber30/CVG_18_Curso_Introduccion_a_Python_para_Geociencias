### Variable names

* Variable names must **start with a letter or the underscore character**, not a number.  
* A variable name can only **contain alpha-numeric characters, and underscores** (A-z, 0-9, and _ ), not special characters. 
* Variable names are **case-sensitive** therefore python, Python, and PYTHON would be different variables.

You should also **avoid naming variables with names that are already used by methods or functions** in the PI or in other libraries.

# Control flow

Control flow structures are basic elements of any programming language. They can control and/or modify the development of programs based on certain conditions or parameters. Here are three advantages of control flow structures in programming:

+ **Automation and efficiency**: They allow different code blocks to execute based on specific conditions, avoiding manual repetition and optimizing the program's flow.
+ **Flexibility and decision-making**: They enable a program to dynamically adapt to different situations, responding to input data and real-time events.
+ **Code clarity and organization**: They help structure programs logically, improving readability and maintainability by dividing the flow into well-defined sections.

The main control flow structures are **conditional statements and loops**, let's review them:

## Conditional statements

A conditional statement is a structure or instruction that allows a program to take a different path, depending if a boolean condition is satisfied or not (False or True). Here is a representation of the conditional statement:

<img src="NB_images/cond_state.png" style="width: 350px;"/>

One consequence of use of the conditional statements is that not all the code in a program is always executed, a portion (Code A or B) will remain untouched. The syntax is as follows :

    if (condition):
        do this for True 
    else:
        do this other thing

Here are some examples of conditional statements, pay attention to the colons(:), brackets, and indentation:

```python
# One condition, one outcome

var1 = 100

if var1 == 100:
    print("The value of the variable is 100")
```

```python
# One condition, two outcomes

a = 2
b = 3

if a == b:
    print("a is equal to b")
else:
    print("a is different to b")
```

```python
# Two conditions, two outcomes

hour = "5 pm"
money = 8

if hour == "5 pm" and money > 10:  
    print("Let's go for some beers")
else:
    print("Keep working!")
```

## Loops

A loop is a structure in which a sequence of instructions is repeated until a condition is satisfied or until a counter reaches a certain value. Here is a representation of the loops:

<img src="NB_images/loop.png" style="width: 250px;"/>

Contrary to the conditional statement after the loops are done (Code A), the other portion of the program (Code B) is executed. There are two types of loops:

* **For loop**
* While loop

Let's see them

### For-loops

For-loops are typically used when the **number of iterations is known or defined** before entering the loop. The Python syntax for a for-loop is:

    for i in iterable:
        Line code 1
        Line code 2
        Line code 3
        .
        .
        .
   
The indentation of the lines (one tab or four spaces) after the `for` statement defines the code to be repeated. For-loops are compact structures because they do not require counters, only an iterable data type. Here are some examples with the function `range` and other iterable items:

```python
# Example 1 - iteration with range

for i in range(6):
    print("Value :", i)
```

```python
# Example 2 - iteration with range

for i in range(3, 11, 2):
    print("Value :", i)
```

```python
# Example 3 - iteration with range and len

var1 = [1, 3, 5, 7, 9]

for i in range(0, len(var1)):
    print("Value :", var1[i])
```

```python
# Example 4 - with iterable item

var1 = [1, 3, 5, 7, 9]

for var in var1:
    print("Value :", var)
```

```python
# Example 5 - with iterable item

text = 'Python'

for letter in text:
    print("Value :", letter)
```

## Local UDF

User-defined functions (UDF) are the base of a more intelligent and efficient (easy to maintain) way of programming. They are pieces of code defined by the user. They are important for the following reasons: 

* They allow you to **separate, organize, and simplify** a large program into smaller parts, making it easier to understand, maintain, and debug.
* When repeated code appears, a function can **group, reuse, and automate** that section so it runs only when the function is called.
* In large projects, developers can **collaborate, divide tasks, and coordinate** their work by creating different functions.

The syntax of a UDF is:

    def function_name(arguments):
        code
        code
        ....
        return values
        
In the UDF the colon (:) and indentation (a tab or four spaces) are critical because they define what is inside the function. Only the return values or variables are known by the rest of the program. Local UDF do not belong to a specific library, they are defined and used (many times if necessary) inside a specific notebook. Here are some examples of local UDF:

```python
# Example of a function with one argument and returning one value

def conversion_C_F(temp_c):
    temp_f = temp_c * 9/5 + 32
    return temp_f
```

## Reading text files

Text files are the most straightforward files used in any computer (regardless of the operative system) to save information in a format familiar to us. Some specific uses of the text files are:

* Simple documents like notes, letters, emails, ...
* Data in columnar shape
* Keepers of parameters or environment variables for programs or operative systems
* Scripts or programs on different programming languages
* Communication between machines, programs, and persons
* Databases

Characters in text files were initially coded for electronic communication (teleprinters) based on different standards that in 1963 became the **ASCII** (American Standard Code for Information Interchange) code. Soon, it was clear that the ASCII code, which only had English characters, some mathematical,  and special symbols (128 characters), was insufficient for other written languages. So the ASCII code started to evolve into different codes until we got to the actual **UTF-8** (Universal Transformation Format) which has 1,112,064 characters, enough for all human written languages and disciplines. The UTF-8 standard is the base on which the World Wide Web works.

More information on character encoding at: https://www.w3schools.com/charsets/

### Reading multi-column text files with missing values

To have a file with just a column is very rare. Normally we have files with several numerical columns, sometimes even columns with text. When you load a multi-column file an array is created with the corresponding number of columns, later each one can be defined as independent variables. Important, to operate with these variables, they have to have the same dimensions (same number of rows).

A common situation is to have several missing or lost values, places where the measurements were not possible. These missing values used to be indicated with special numbers such as -999.000, or -999.2500, or just with empty spaces. Numpy recognizes automatically empty spaces as missing values and replaces them with **nan**, not a number.

We are going to load the *rampi.csv* file, also in the Input folder, which is a five-column file that comes from a study of gold and associated elements (in ppm) in quartz veins in the **Rampi block prospect**, Indonesia (see image below from Google Earth):

```python
# Let's plot the concentration of zinc against its real location along the vein, each 2 m

# This will be a plot with two real x-axis

x = range(0, 2*len(zn), 2)

plt.figure(figsize=(8, 5))

plt.plot(x, zn)

plt.xlabel("Distance (m)")
plt.ylabel("Weight %")
plt.title("Zn along quartz vein")
plt.show()
```

```python
# Let's show the mean value in the previous plot

y_mean = np.mean(zn) # y is a single value

plt.figure(figsize=(8,5))
plt.plot(x, zn)

# axhline draws a horizontal line based on y

plt.axhline(y_mean, ls="-", color='red', label="mean")

plt.legend()

plt.xlabel("Distance (m)")
plt.ylabel("Weight %")
plt.title("Zn along quartz vein")
plt.show()
```

```python
# Let's fill between the mean and the curve

plt.figure(figsize=(8,5))
plt.plot(x, zn)
plt.axhline(y_mean, ls="-", color='black', label="mean")

plt.fill_between(x, zn, y_mean, where=(zn >= y_mean), facecolor='b', interpolate=True, label="over mean")
plt.fill_between(x, zn, y_mean, where=(zn <= y_mean), facecolor='r', interpolate=True, label="under mean")

plt.xlabel("Distance (m)")
plt.ylabel("Weight %")
plt.title("Zn along quartz vein")
plt.legend()
plt.show()
```

## Multiplots by subplot

With the Matploplib's **subplot** function it is possible to gather several plots in a composition, arranged based on a rectangular grid of raw and columns. The arguments (2,2,1) in the following command:

`plt.subplot(2,2,1)`
    
means the first plot in an arrange of two rows and two columns of plots, something like this:

<img src="NB_images\multiplot_example.png"  style="width:700px">

```python
# Building and saving a 2 x 3 multiplot with the Rampi data

plt.figure(figsize=(16,10))

plt.subplot(2,3,1)

# Point-line plot
plt.plot(au, 'y', label="Au")
plt.plot(ag, 'c', label="Ag")
plt.plot(cu, 'r', label="Cu")
plt.plot(zn2,'m', label="Zn")
plt.plot(pb, 'b', label="Pb")
plt.title("Elements")
plt.xlabel("File order")
plt.ylabel("ppm")
plt.legend()
plt.grid(True)

plt.subplot(2,3,2)

# Bar plot
elements = ('Au', 'Ag', 'Cu', 'Zn', 'Pb')
sizes = [np.nanmean(au), np.nanmean(ag), np.nanmean(cu), np.nanmean(zn2), np.nanmean(pb)]
x = np.arange(len(elements))   
colors = ['y', 'c', 'r','m','b']
plt.bar(x, sizes, align='center', alpha=0.5, color=colors)
plt.xticks(x, elements)
plt.title('Elements')
plt.ylabel('ppm')
plt.xlabel('Elements')
plt.grid(True)

plt.subplot(2,3,3)

# Cross plot
plt.loglog(au,ag,'o')
plt.title("Au vs. Ag")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Ag (ppm)")
plt.grid(True)

plt.subplot(2,3,4)

# Cross plot
plt.loglog(au,cu,'o')
plt.title("Au vs. Cu")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Cu (ppm)")
plt.ylim(1,200)
plt.grid(True)

plt.subplot(2,3,5)

# Cross plot
plt.loglog(au,zn2,'o')
plt.title("Au vs. Zn")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Zn (ppm)")
plt.ylim(1,200)
plt.grid(True)

plt.subplot(2,3,6)

# Cross plot
plt.loglog(au,pb,'o')
plt.title("Au vs. Pb")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Pb (ppm)")
plt.ylim(1,200)
plt.grid(True)

# Avoid the overlap between labels

plt.tight_layout()

# Saving multiplot to file

plt.savefig("Output/multiplot.png")
plt.savefig("Output/multiplot.pdf")

plt.show()
```

## Multiplot by looping an array

Multiplots can be built by a loop, making the code shorter and more efficient. Let's see an example done by slicing the Rampi array: 

```python
rampi.shape
```

```python
# A Unix command

%more Input/rampi.csv
```

```python
# This in gold

rampi[:,0]
```

```python
# Elements, colors, and position in the vein, according to the rampi array

elements = ['Au', 'Ag', 'Cu', 'Zn', 'Pb']
colors = ['y', 'c', 'r', 'm', 'b']
x = range(0, 3*rampi.shape[0], 3)

# Loop according to the number of columns in the rampi array

for i in range(rampi.shape[1]):
    
    plt.figure(figsize=(16,4))
    
    # Line plot of the element
    plt.subplot(1,2,1)
    plt.plot(x, rampi[:,i], color=colors[i], alpha=0.5)
    plt.xlim(0, 65)
    plt.xlabel('Position in the vein (m)')
    plt.ylabel('Concentration (ppm)')
    plt.title('Plot of '+elements[i])
    plt.grid()
    
    # Histogram of the element
    plt.subplot(1,2,2)
    plt.hist(rampi[:,i], color=colors[i], alpha=0.5)
    plt.ylabel('Frequency')
    plt.xlabel('Concentration (ppm)')
    plt.title('Histogram of '+elements[i])
    plt.grid()
    plt.savefig("Output/plot_"+elements[i]+".png")  
     
plt.tight_layout()
plt.show()        
```
